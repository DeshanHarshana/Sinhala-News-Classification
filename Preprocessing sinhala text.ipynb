```python
import nltk
nltk.download('punkt')
```

```python
from sinling import SinhalaTokenizer

tokenizer = SinhalaTokenizer()

sentence = 'අමරදේවයන් විසින් රචනා කරන ලද කලේද?' 

tokenizer.tokenize(sentence)
```

```python

from sinling import SinhalaStemmer
stemmer = SinhalaStemmer()

s='අමරදේවයන් විසින් රචනා කරන ලද කලේද?'
word_list=tokenizer.tokenize(s)

print([word_list])

for word in word_list:
  lst=stemmer.stem(word)
  print(lst)

#print(word_list)
#stemmer.stem(word)
```
